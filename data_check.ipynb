```python
import pandas as pd

# データの読み込み
df = pd.read_csv("combined_base_data.csv")

# 1. 欠損値の確認（どこが空っぽか？）
print("--- 欠損値の数 ---")
print(df.isnull().sum())

# 2. 価格が0円または空のものを抽出して表示
missing_price = df[df['価格'].isnull() | (df['価格'] == 0)]
print(f"\n--- 価格に問題がある商品: {len(missing_price)}件 ---")
display(missing_price.head()) # displayを使うと表形式で綺麗に出ます

# 3. ショップごとの取得件数確認
print("\n--- ショップ別データ件数 ---")
print(df['ショップ名'].value_counts())
```
